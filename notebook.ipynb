```python
sns.boxplot(x=df["State_Name"],y=df["Production"])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x=df["District_Name"],y=df["Production"])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x=df["Crop_Year"],y=df["Production"])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x=df["Season"],y=df["Production"])
```

```python
sns.boxplot(x=df["Crop"],y=df["Production"])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.tick_params(labelsize=10)
sns.heatmap(df.corr(),annot=True)
```

```python
df["State_Name"].unique()
```

```python
plt.tick_params(labelsize=10)
df.groupby("Crop_Year")["Production"].agg("sum").plot.bar()
```

```python
plt.figure(figsize=(25,8))
plt.tick_params(labelsize=10)
df.groupby("Crop")["Production"].agg("count").plot.bar()
plt.show()
```

```python
sns.countplot(x ='Season', data = df, palette ='plasma')
```

```python
sns.barplot(x='Season',y='Production',data=df)
```

```python
plt.figure(figsize=(20,30))

ax = sns.countplot(y="Crop", data=df)

plt.setp(ax.get_xticklabels(), rotation=0,fontsize=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(22,30)) 

ax = sns.barplot(y=df[(df['Crop']!='Coconut') & (df['Crop']!='Sugarcane') & (df['Crop']!='Banana')]['Crop'],x=df['Area'])
plt.setp(ax.get_xticklabels(), rotation=0,fontsize=20)
plt.tight_layout()
plt.show()
```
